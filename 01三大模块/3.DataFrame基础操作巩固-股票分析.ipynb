### 需求：股票分析
- 使用tushare包获取某股票的历史行情数据。
- 输出该股票所有收盘比开盘上涨3%以上的日期。
- 输出该股票所有开盘比前日收盘跌幅超过2%的日期。
- 假如我从2010年1月1日开始，每月第一个交易日买入1手股票，每年最后一个交易日卖出所有股票，到今天为止，我的收益如何？

```python
data = ts.get_k_data(code='600519',start='1999-1-1')
data
```

```python
data.info()
#发现date列为字符串类型，需要将其转换成时间序列类型
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data.info()
```

```python
data.head()
```

```python
#将date列的数据作为整个源数据的行索引
data.set_index('date',inplace=True)
```

```python
data.head()
```

```python
#输出该股票所有收盘比开盘上涨3%以上的日期
#(收盘-开盘)/开盘 > 0.03
(data['close'] - data['open']) / data['open'] > 0.03
```

```python
#经验分享：在df的相关处理中，如果获取了一组布尔值，下一步马上将布尔值作为df的行索引即可
data.loc[(data['close'] - data['open']) / data['open'] > 0.03]
#如果将布尔值作为源数据的行索引，表示将true对应的行数据保留，false对应的行数据舍弃
```

```python
data.loc[(data['close'] - data['open']) / data['open'] > 0.03].index
```

```python
#输出该股票所有开盘比前日收盘跌幅超过2%的日期。
#(开盘-前日收盘)/前日收盘 < -0.02
(data['open'] - data['close'].shift(1))/data['close'].shift(1) < -0.02
```

```python
#获取了满足需求的行数据
data.loc[(data['open'] - data['close'].shift(1))/data['close'].shift(1) < -0.02]
```

```python
data.loc[(data['open'] - data['close'].shift(1))/data['close'].shift(1) < -0.02].index
```

- 假如我从2010年1月1日开始，每月第一个交易日买入1手股票，每年最后一个交易日卖出所有股票，到今天为止，我的收益如何？
- 分析：买入股票使用开盘价，卖出股票使用收盘价
    - 买
        - 一个完整的年，需要买入12次股票，每次买入一手，共买入1200只股票
        - 买股票的单价，就是每次买入一手股票日期的开盘价。
    - 卖
        - 一个完整的年，需要卖出一次股票，一次卖出1200只。
        - 卖出股票的单价就是卖股票当天的收盘价作为单价
    - 特殊情况：
        - 2020年不是一个完整的年，只可以买入股票，无法卖出股票。但是手中剩余的股票价值也是需要计算到总收益中。

```python
df = data['2010-1':'2020-9']
```

```python
#需要获取买入股票的股票单价
df_monthly = df.resample('M').first() #获取了每一年每一个月的第一个交易日对应的行数据
df_monthly.head()
```

```python
#买股票花的钱
cost_money = df_monthly['open'].sum()*100
cost_money
```

```python
#卖股票
df_yearly = df.resample('A').last()[:-1] #每一年最后一个交易日对应的行数据
df_yearly
```

```python
recv_money = df_yearly['close'].sum() * 1200
recv_money
```

```python
#用最后一个交易日的收盘价作为剩余股票价值的单价
last_money = df.iloc[-1]['close'] * 900
```

```python
#总收益
last_money + recv_money - cost_money
```
