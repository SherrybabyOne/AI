### plt.plot()绘制线性图
- 绘制单条线形图
- 绘制多条线形图
- 设置坐标系的比例plt.figure(figsize=(a,b)) 
- 设置图例legend()
- 设置轴的标识
- 图例保存
    - fig = plt.figure()
    - plt.plot(x,y)
    - figure.savefig（）
- 曲线的样式和风格（自学）

```python
x = np.linspace(-np.pi,np.pi,num=10)
y = 2*x + 0.123
```

```python
plt.plot(x,y)
```

```python
x = x
y = x**2
plt.plot(x,y)
```

```python
#在一个坐标系中绘制多条曲线
plt.plot(x,y,x+1,y-1)
```

```python
plt.plot(x,y)
plt.plot(x**2,y**0.5)
```

```python
#设置坐标系的比例
plt.figure(figsize=(5,5))
plt.plot(x,y)
```

```python
#曲线设置图例
plt.plot(x,y,label='temp')
plt.plot(x+2,y-1,label='dist')
plt.legend(loc=9,ncol=1)
```

```python
#图片的保存
fig = plt.figure() #step1
#step2:绘图
plt.plot(x,y,label='temp')
plt.plot(x+2,y-1,label='dist')
plt.legend(loc=9,ncol=1)
#step3：保存
fig.savefig('./123.png')
```

### 直方图
- 是一个特殊的柱状图，又叫做密度图
- plt.hist()的参数
    + bins  
    可以是一个bin数量的整数值，也可以是表示bin的一个序列。默认值为10
    + normed  
    如果值为True，直方图的值将进行归一化处理，形成概率密度，默认值为False
    + color  
    指定直方图的颜色。可以是单一颜色值或颜色的序列。如果指定了多个数据集合,例如DataFrame对象，颜色序列将会设置为相同的顺序。如果未指定，将会使用一个默认的线条颜色
    + orientation  
    通过设置orientation为horizontal创建水平直方图。默认值为vertical

```python
x = [1,1,1,2,3,4,5,5,5,6,7,8,8]
plt.hist(x,bins=15)
```

### 饼图
- pie()，饼图也只有一个参数x
- 饼图适合展示各部分占总体的比例，条形图适合比较各部分的大小

```python
arr=[11,22,31,15]
plt.pie(arr)
```

```python
arr=[0.2,0.3,0.1]
plt.pie(arr)
```

```python
arr=[11,22,31,15]
plt.pie(arr,labels=['a','b','c','d'])
```

```python
arr=[11,22,31,15]
plt.pie(arr,labels=['a','b','c','d'],labeldistance=0.3)
```

```python
arr=[11,22,31,15]
plt.pie(arr,labels=['a','b','c','d'],labeldistance=0.3,autopct='%.6f%%')
```

```python
arr=[11,22,31,15]
plt.pie(arr,labels=['a','b','c','d'],labeldistance=0.3,shadow=True,explode=[0.2,0.3,0.2,0.4])
```

```python
x = np.linspace(-np.pi,np.pi,num=30)
y = (x ** 2) / 0.5**-2
```

```python
plt.scatter(x,y)
```
